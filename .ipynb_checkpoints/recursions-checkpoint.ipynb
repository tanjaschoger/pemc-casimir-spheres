# Modified Bessel functions

## Half-integer order 

\begin{align}
I_{n-3/2}(x)  &= \frac{2n-1}{x} I_{n-1/2}(x) + I_{n+1/2} \\
K_{n+1/2}(x)  &= \frac{2n-1}{x} K_{n-1/2}(x) + K_{n-3/2}(x) \\
\end{align}

### small arguments

\begin{align}
I_{n+1/2}(x) &=   \frac{x^{n +1/2}2^{n+1} (n+1)!}{(2n+2)!} \tilde{I}_{n+1/2} (x) \\
K_{n+1/2}(x) &=   \frac{(2n)!}{x^{n +1/2}2^{n} (n)!} \tilde{K}_{n+1/2} (x)
\end{align}

Recursion relations for the scaled Bessel functions

\begin{align}
\tilde{I}_{n-3/2}(x) &= \tilde{I}_{n-1/2}(x) + \frac{x^2}{4n^2 -1} \tilde{I}_{n+1/2}(x) \\
\tilde{K}_{n+1/2}(x) &= \tilde{K}_{n-1/2}(x) + \frac{x^2}{(2n-1)(2n-3)} \tilde{K}_{n-3/2}(x)
\end{align}

### large arguments

\begin{align}
I_{n+1/2}(x) &= e^{x(\sqrt{1+\Lambda^2} - \Lambda \mathrm{arcsinh}(\Lambda))}\tilde{I}_{n+1/2}(x) \\
K_{n+1/2}(x) &= e^{-x(\sqrt{1+\Lambda^2} - \Lambda \mathrm{arcsinh}(\Lambda))}\tilde{K}_{n+1/2}(x)
\end{align}
where 
$\Lambda = \frac{n+1/2}{x}$

## Integer orders


### large arguments

\begin{equation}
I_n (x) = e^{x}\tilde{I}_n(x)
\end{equation}
